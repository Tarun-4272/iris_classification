```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("iris intern.csv")
```

```python
df
```

```python
df.isnull()
```

```python
pd.get_dummies(df['species'])
```

```python
df['species']
```

```python
species=pd.get_dummies(df['species'])
```

```python
df
```

```python
df=pd.concat([df,species],axis=1)
```

```python
df
```

```python
df=df.drop('species',axis=1)
```

```python
df
```

```python
x=df.drop(['Iris-setosa','Iris-versicolor','Iris-virginica'],axis=1)
```

```python
x
```

```python
y=df[['Iris-setosa','Iris-versicolor','Iris-virginica']]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=30,random_state=42)
```

```python
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
clf=KNeighborsClassifier(n_neighbors=3)
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
y_pred
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```
